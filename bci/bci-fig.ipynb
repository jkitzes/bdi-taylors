This notebook reads the csv saved by bci-calcs and makes a density plot. It should be run using an R kernel.

```r
z_table <- read.csv("bci_zs.csv")
```

```r
head(z_table)
```

```r
plot_z_density <- function(){
    plot(density(z_table[,"z"]), main="", xlab="z", bty="n", xlim=c(-1,1.5))
    lines(c(0.5,0.5), c(-.1,2.25), col="blue", lty=5)
    }
plot_z_density()
```

```r
pdf("z_density.pdf", pointsize=10, width=4, height=4)
    p <- plot_z_density()
    print(p)
dev.off()
```

```r
plot_c_density <- function(){
    plot(density(z_table[,"c"]), main="", xlab="c", bty="n", xlim=c(-15,10))
    }
plot_c_density()
```

```r
pdf("c_density.pdf", pointsize=10, width=4, height=4)
    p <- plot_c_density()
    print(p)
dev.off()
```
