This notebook reads a csv containing Xiao's spatial Taylor's law values of `b` and makes a density plot.

The data in `spatial_TL_data.xlsx` were received in an email from Xiao Xiao. The key column is `observed_b`, as described in the Supplementary Information, and the `type` column which allows us to extract only those reflecting spatial (not temporal) patterns.

```r
library(openxlsx)
```

```r
tl_data <- read.xlsx("spatial_TL_data.xlsx")
```

```r
head(tl_data)
```

```r
tl_data_spatial <- tl_data[tl_data["spatial"]==1,]
```

```r
# Now we have only the 45 rows of spatial data, with all 7 columns
dim(tl_data_spatial)
```

```r
plot_density <- function(){
    plot(density(tl_data_spatial[,"observed_b"]), main="", xlab="b", bty="n", xlim=c(-1,4))
    lines(c(1.5,1.5), c(-.1,1.2), col="red", lty=5)
    }
plot_density()
```

```r
pdf("density.pdf", pointsize=10, width=5, height=5)
    p <- plot_density()
    print(p)
dev.off()
```
